```python
import pandas as pd
```

```python
df = data = {
    "Size": ["S", "M", "L", "XL", "S", "M",'2XL','S']
}
```

```python
dt = pd.DataFrame(df)
```

```python
dt
```

```python
oe_enc = [['S','M','L','XL','2XL']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enc = OrdinalEncoder(categories=oe_enc)
dt['dt_enco']=enc.fit_transform(dt[['Size']])
```
